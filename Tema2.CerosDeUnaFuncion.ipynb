```python
# 1.2. Programa para determinar si el resultado de multiplicar los extremos del intervalo 
#    es positivo o negativo (todavía no esta terminado en este punto)

def function(x):
    return (x**3) + (1.71 * (x**2)) - (26.21*x) - 44.82

delta_x = 0.2
start = 4.0
end = 6.0

x1 = start
x2 = start + delta_x

while x2 <= end:
    print(f"Intervalo: {x1} , {x2}",end=" ")
    
    f1 = function(x1)
    f1 = round(f1,4)
    f2 = function(x2)
    f2 = round(f2,4)
    print(f"f(x) en los extremos: {f1} , {f2}",end= " ")
    
    if (f1 * f2) <= 0:
        print("Hubo cambio de signo!")
    else:
        print("No hubo cambio de signo")
        
    x1 = x1 + delta_x
    x1 = round(x1,4)
    x2 = x2 + delta_x
    x2 = round(x2,4)
```

```python
# 1.3. Programa basado en el anterior (punto 2), pero en esta ocasión ya se teniene 
#    la ejecución cuando se encuentra un cambio de signo (se detecta una raíz)

def incremental_search_method(start,end,delta_x):
    x1 = start
    x2 = start + delta_x
    
    while x2 <= end:
        print(f"Intervalo: {x1} , {x2}",end=" ")
    
        f1 = function(x1)
        f1 = round(f1,4)
        f2 = function(x2)
        f2 = round(f2,4)
        print(f"f(x) en los extremos: {f1} , {f2}",end= " ")
    
        if (f1 * f2) <= 0:
            print("Hubo cambio de signo!")
            return x1,x2    # Se agrega el return para terminar la ejecución en cuanto detecte cambio de signo
        else:
            print("No hubo cambio de signo")
        
        x1 = x1 + delta_x
        x1 = round(x1,4)
        x2 = x2 + delta_x
        x2 = round(x2,4)
    
    return 0,0           # Este return se agrega para que en el caso de que no se encuentre
                         # alguna raíz indique al usuario su ausencia
a,b = incremental_search_method(4.0,6.0,0.2)
print(f"\nLa raíz esta en el intervalo: ({a} , {b})")
```

```python
# 2. Método de bisección

#Grafica para visualizar en donde se encuentran las raíces 

import numpy as np
import matplotlib.pyplot as plt

x = np.arange(-10.0,10.0,0.01)
y = (x**3) + (1.71 * (x**2)) - (26.21*x) - 44.82

plt.figure(figsize = (14,8))
plt.plot(x,y)

plt.axhline(0, color = 'red')
plt.axvline(0, color = 'green')
plt.show()
```

```python
# 2. Método de la bisección (se usará la misma funcion que el ejemplo de búsqueda incremental)
#    Se detendra el algoritmo por medio de controlar el número de iteraciones

def f(x):
    return (x**3) + (1.71 * (x**2)) - (26.21*x) - 44.82

def bisection_method(function, start, end, iterations):
    x1 = start
    x2 = end
    midpoint = (x1 + x2) / 2
    
    for i in range(0, iterations):
        f1 = function(x1)
        f2 = function(x2)
        fm = function(midpoint)
        
        print(f"Iteración {i+1}")
        print(f"Intervalo: ({x1} , {x2}), \nPunto medio: {midpoint}")
        print(f"f(x) en el intervalo: \n\tIzquierda {f1}, \n\tCentro {fm}, \n\tDerecha {f2}")
        
        if(fm == 0):
            return midpoint   # Este return significa que la raiz se encuentra en ese punto medio
        else:
            if(f1*fm)<0:
                print("Elijo el intervalo izquierdo\n")
                x2 = midpoint
            else:
                print("Elijo el intervalo derecho\n")
                x1 = midpoint
        
        midpoint = (x1 + x2) / 2
        
    return midpoint   # Este return indica que una vez realizadas todas las iteraciones se
                      # se considerará el punto medio como respuesta
    
x = bisection_method(f,4.0,6.0,30)
print(f"La raiz esta en: {x}")
print(f"f(x) = {f(x)}")
```

```python
# 2.1. Version final del programa puliendo el programa eliminando los print que no son necesarios

def f(x):
    return (x**3) + (1.71 * (x**2)) - (26.21*x) - 44.82

def bisection_method(function, start, end, iterations):
    x1 = start
    x2 = end
    midpoint = (x1 + x2) / 2
    
    for i in range(0, iterations):
        f1 = function(x1)
        f2 = function(x2)
        fm = function(midpoint)
       
        if(fm == 0):
            return midpoint   
        else:
            if(f1*fm)<0:
                x2 = midpoint
            else:
                x1 = midpoint
        
        midpoint = (x1 + x2) / 2
        
    return midpoint   
    
x = bisection_method(f,-6.0,-4.0,30)   # Calculando distintas raíces
print(f"La raiz x1 esta en: {x}")
print(f"f(x1) = {f(x)}")

x = bisection_method(f,-0.5,-2.5,30)
print(f"\nLa raiz x2 esta en: {x}")
print(f"f(x2) = {f(x)}")

x = bisection_method(f,4.0,6.0,30)
print(f"\nLa raiz x3 esta en: {x}")
print(f"f(x3) = {f(x)}")
```
