### Como graficar un segmento

- Se deben de poner los intervalos de los ejes *x* y *y*. El primer argumento del ejemplo nos dice que se desea una recta que en el *x* vaya de 3 a 6, mientras que en el eje *y* vaya de 2 a 8. Una forma f√°cil de entender esto, es pensar que si queremos un segmento de recta que vaya del punto P1(*x1* , *y1*) al punto P1(*x2* , *y2*) entonces deberiamos usar plt.show([*X1* , *X2*] , [*y1* , *y2*]).

```python
# Ejemplo

import matplotlib.pyplot as plt
plt.plot([3,6] , [2,8])
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.plot([1,8] , [-4,6] , marker = 'o' , color = 'red')  # marker es para indicar que se quiere tener en los extremos del segmento
plt.plot([-1,5] , [2,-4] , marker = '.' , color = 'blue')  # color es ara indicar el color del segmento
plt.plot([7,-2] , [-6,8] , marker = '+' , color = 'green')
plt.plot([6,6] , [6,6] , marker = 'x' , color = 'purple')

plt.plot()
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize = (8,8))  # Indica el tama√±o de la figura
plt.axis([-10,10,-10,10])    # Indica de donde a donde es la medida de mis ejes x y y
plt.axhline(0, color = 'gray')  # Aqu√≠ se traza el eje x
plt.axvline(0, color = 'gray')  # Aqu√≠ se traza el eje y

plt.plot([1,8] , [-4,6] , marker = 'o' , color = 'red')  
plt.plot([-1,5] , [2,-4] , marker = '.' , color = 'blue')
plt.plot([7,-2] , [-6,8] , marker = '+' , color = 'green')
plt.plot([6,6] , [6,6] , marker = 'x' , color = 'purple')

plt.plot()
plt.show()
```

```python
# Dibujando al gato

import matplotlib.pyplot as plt

Points = {'A':[1,7], 'B':[2,6], 'C':[3,6], 'D':[4,7], 'E':[4,5], 'F':[9,5], 'G':[9,8], 'H':[11,8],
          'I':[11,7], 'J':[10,7], 'K':[10,4], 'L':[8,4], 'M':[8,1], 'N':[6,1], 'O':[6,2],
          'P':[7,2], 'Q':[7,3], 'R':[4,3], 'S':[4,1], 'T':[2,1], 'U':[2,2], 'V':[3,2],
          'W':[3,4], 'X':[1,4], 'Y':[2,5], 'Z':[3,5]}                               # Indica los puntos de la figura

Lines = ['AB', 'BC', 'CD', 'DE', 'EF', 'FG', 'GH', 'HI', 'IJ', 'JK', 'KL', 'LM', 'MN', 'NO',
         'OP', 'PQ', 'QR', 'RS', 'ST', 'TU', 'UV', 'VW', 'WX', 'XA', 'YY', 'ZZ']    # Indica que punto con que punto se quiere conectar

for line in Lines:
  P1 = line[0]
  P2 = line[1]
  x,y = [Points[P1][0], Points[P2][0]] , [Points[P1][1], Points[P2][1]]
  plt.plot(x, y, marker = 'o', color = 'blue')

plt.show()
```

```python
# Ejemplo de transformaciones lineales

import matplotlib.pyplot as plt
import math

plt.figure(figsize = [8,8])
plt.axis([-1, 20, -1, 20])
plt.axhline(0, color = 'gray')
plt.axvline(0, color = 'gray')

Points = {'A':[1,7], 'B':[2,6], 'C':[3,6], 'D':[4,7], 'E':[4,5], 'F':[9,5], 'G':[9,8], 'H':[11,8],
          'I':[11,7], 'J':[10,7], 'K':[10,4], 'L':[8,4], 'M':[8,1], 'N':[6,1], 'O':[6,2],
          'P':[7,2], 'Q':[7,3], 'R':[4,3], 'S':[4,1], 'T':[2,1], 'U':[2,2], 'V':[3,2],
          'W':[3,4], 'X':[1,4], 'Y':[2,5], 'Z':[3,5]}                               # Indica los puntos de la figura

Lines = ['AB', 'BC', 'CD', 'DE', 'EF', 'FG', 'GH', 'HI', 'IJ', 'JK', 'KL', 'LM', 'MN', 'NO',
         'OP', 'PQ', 'QR', 'RS', 'ST', 'TU', 'UV', 'VW', 'WX', 'XA', 'YY', 'ZZ']    # Indica que punto con que punto se quiere conectar

T = [[0,1] , [1,0]]       # Esta es la matriz

for i in Points:                       # Aplicando transformacion (multiplicando la matriz por el vector)
  v = [Points[i][0], Points[i][1]]
  Points[i][0] = (T[0][0] * v[0]) + (T[0][1] * v[1])
  Points[i][1] = (T[1][0] * v[0]) + (T[1][1] * v[1])

for line in Lines:
  P1 = line[0]
  P2 = line[1]
  x,y = [Points[P1][0], Points[P2][0]] , [Points[P1][1], Points[P2][1]]
  if P1 != P2:
    plt.plot(x, y, color = 'black')
  else:
    plt.plot(x, y, marker = 'o', color = 'black')

plt.show()
```

```python
# Ejemplo de transformaciones lineales (rotacion)

import matplotlib.pyplot as plt
import math

plt.figure(figsize = [8,8])
plt.axis([-10, 20, -10, 20])
plt.axhline(0, color = 'gray')
plt.axvline(0, color = 'gray')

Points = {'A':[1,7], 'B':[2,6], 'C':[3,6], 'D':[4,7], 'E':[4,5], 'F':[9,5], 'G':[9,8], 'H':[11,8],
          'I':[11,7], 'J':[10,7], 'K':[10,4], 'L':[8,4], 'M':[8,1], 'N':[6,1], 'O':[6,2],
          'P':[7,2], 'Q':[7,3], 'R':[4,3], 'S':[4,1], 'T':[2,1], 'U':[2,2], 'V':[3,2],
          'W':[3,4], 'X':[1,4], 'Y':[2,5], 'Z':[3,5]}                               # Indica los puntos de la figura

Lines = ['AB', 'BC', 'CD', 'DE', 'EF', 'FG', 'GH', 'HI', 'IJ', 'JK', 'KL', 'LM', 'MN', 'NO',
         'OP', 'PQ', 'QR', 'RS', 'ST', 'TU', 'UV', 'VW', 'WX', 'XA', 'YY', 'ZZ']    # Indica que punto con que punto se quiere conectar

radianes = round(math.radians(45), 4)

T = [[math.cos(radianes), -math.sin(radianes)] , [math.sin(radianes), math.cos(radianes)]]

for i in Points:
  v = [Points[i][0], Points[i][1]]
  Points[i][0] = (T[0][0] * v[0]) + (T[0][1] * v[1])
  Points[i][1] = (T[1][0] * v[0]) + (T[1][1] * v[1])

for line in Lines:
  P1 = line[0]
  P2 = line[1]
  x,y = [Points[P1][0], Points[P2][0]] , [Points[P1][1], Points[P2][1]]
  if P1 != P2:
    plt.plot(x, y, color = 'black')
  else:
    plt.plot(x, y, marker = 'o', color = 'black')

plt.show()
```
